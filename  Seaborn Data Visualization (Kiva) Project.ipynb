```python

VISUALIZING LOANS AWARDED BY KIVA

The project uses Seaborn to explore the average loan amount by country using aggregated bar charts.
The project also visualizes the distribution of loan amounts by project type and gender using box plots and violin plots.
```

```python
#Importing the necessary modules

from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
Overview of the Dataset:

Each entry (row) in the dataset represents a loan that Kiva awarded to a particular project.
The loan_amount column shows the amount (in U.S. dollars) awarded to the project. 
The activity column has the category type that the project falls under. 
The country column is the country where the project is located. 
The gender column represents the gender of the primary person who applied for the loan.
```

```python
# Data Examination
df = pd.read_csv('kiva_data.csv')
print(df.head(25))
 
```

```python
# Creating Bar Plot
# The bar plot  visualizes the average size of Kiva loans given to projects, by country.

import matplotlib.ticker as mtick

f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(data=df, x="country", y = "loan_amount")
ax.set_title("Visualizing Loan Amount By Country")


fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

```python
# Creating Bar Plot by Category
# The bar plot by visualizes  the loan amount by country, and "nesting" by gender

sns.set_palette("Dark2")
sns.set_style("darkgrid")
f, ax = plt.subplots(figsize=(25, 15))
sns.barplot(data=df, x="country", y="loan_amount", hue = "gender")
ax.set_title("Visualizing Loan Amount By Country")


```

```python
From the above bar plot, the following deductions are made:
    
Female recipients receive lower loans compared to the male recipients
El Salvador has the least disparity in loan amounts awarded by gender.
```

```python
#creating Box Plots
# The box plot to compares the distribution of loans by country.

sns.set_palette("Accent")
sns.set_style("darkgrid")
plt.figure(figsize=(16, 10))
sns.boxplot(data=df, x="country", y="loan_amount")
ax.set_title("Visualizing Loan Amount By Country")
```

```python
The box plot above shows that Kenya has the highest Loan Distribution while  Cambodia offers the highest loan amount.
```

```python
#creating Box Plots
# The box plot to compares the distribution of loans by activity.
# It shows you the loan amounts by activity, rather than country. 
# On average, "Farming" receives the largest loan amount and also has the widest distribution.

sns.set_palette("Accent")
sns.set_style("darkgrid")
plt.figure(figsize=(16, 10))
sns.boxplot(data=df, x="activity", y="loan_amount")
ax.set_title("Visualizing Loan Amount By Country")
```

```python
#creating Violin Plots
# The violin  plot  compares the distribution of loans by activity.

sns.set_palette("Pastel2")
sns.set_style("whitegrid")
plt.figure(figsize=(16, 10))
sns.violinplot(data=df, x="activity", y="loan_amount")
plt.title("Visualizing Loan Amount By Activity")
```

```python
#creating Violin Plots
# The violin plot  compares the distribution of loans by Country.

sns.set_palette("Set3")
sns.set_style("whitegrid")
plt.figure(figsize=(16, 10))
sns.violinplot(data=df, x="country", y="loan_amount")
plt.title("Visualizing Loan Amount By Country")
```

```python
#creating Violin Plots
# The violin plot compares the distribution of loans by country and "nesting" by gender.

sns.set_palette("Spectral")
sns.set_style("darkgrid")
plt.figure(figsize=(18, 12))
sns.violinplot(data=df, x="country", y="loan_amount", hue = "gender", split = True)
plt.title("Visualizing Loan Amount By Country")
```
